Reuse the kNN-VC codes to extract speech representations for both source audio and target audio

```python
import torch, torchaudio
knn_vc = torch.hub.load('bshall/knn-vc', 'knn_vc', prematched=True, trust_repo=True, pretrained=True, device='cuda')
# path to 16kHz, single-channel, source waveform
src_wav_path = '/content/Phoneme_Hallucinator/audio/source.wav'
# list of paths to all reference waveforms (each must be 16kHz, single-channel) from the target speaker
ref_wav_paths = ['/content/Phoneme_Hallucinator/audio/target.wav', ]
src_audio, sr = torchaudio.load(src_wav_path)

ref_audio, sr = torchaudio.load(ref_wav_paths[0])

query_seq = knn_vc.get_features(src_wav_path)
matching_set = knn_vc.get_matching_set(ref_wav_paths)
unexpanded_out_wav = knn_vc.match(query_seq, matching_set, topk=4)

torch.save(matching_set, "matching_set.pt")
```

Perform set expansion using our Phoneme Hallucinator. You can change how many new representations to hallucinate by changing "--num_samples".

```python
%cd Phoneme_Hallucinator
!python ./scripts/speech_expansion_ins.py --cfg_file=./exp/speech_XXL_cond/params.json --num_samples 15000 --path /content/matching_set.pt --out_path /content/expanded_set.npy
%cd ..
import numpy as np
import IPython
expanded_set = torch.tensor(np.load("expanded_set.npy")).to(matching_set.device)
matching_set = torch.cat([matching_set, expanded_set], 0)
out_wav = knn_vc.match(query_seq, matching_set, topk=4)
```

Listen to audio results. "unexpanded_converted" is the result of kNN-VC. "converted" is our result.

```python
import IPython.display as ipd
print("source")
IPython.display.display(ipd.Audio(src_audio.numpy(), rate=sr))
print("target")
IPython.display.display(ipd.Audio(ref_audio.numpy(), rate=sr))
print("unexpanded_converted")
IPython.display.display(ipd.Audio(unexpanded_out_wav.numpy(), rate=16000))
print("converted")
IPython.display.display(ipd.Audio(out_wav.numpy(), rate=16000))
```
